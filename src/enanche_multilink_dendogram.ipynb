```python
import pandas as pd
```

```python
dendo = pd.read_excel("../data/dendogram_multilink_full.xlsx", header=None, names=["c1","c2","dist"])
```

```python
dendo.head()
```

```python
n_data = 230
counter = dict()
point_count = []

for i in range(n_data+1):
    counter[i] = 1

for index, row in dendo.iterrows():
    counter[n_data +1  + index] = counter[row["c1"]] + counter[row["c2"]] 
    point_count.append(counter[n_data +1  + index])

point_count
```

```python
dendo["n_points"] = point_count
```

```python
dendo.to_excel("../data/dendogram_enanched.xlsx")
```
